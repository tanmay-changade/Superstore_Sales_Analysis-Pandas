```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Sample - Superstore.csv",encoding = 'latin-1')
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

converting date columns

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data.head()
```

```python
data.info()
```

```python
data['Order Month'] = data['Order Date'].dt.month
data['Order Year'] = data['Order Date'].dt.year
data['Order Day of Week'] = data['Order Date'].dt.dayofweek
```

calculate the monthly sales of the store and identify which month had the highest sales and which month had the lowest sales

```python
sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
fig = plt.plot(sales_by_month['Order Month'],sales_by_month['Sales'],marker = 'o')
plt.xlabel("Order Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Analysis")
plt.grid(alpha = 0.2)
plt.show()
```

# analyze sales based on product categories and determine which category has the lowest sales and which category has the highest sales.

```python
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
```

```python
sales_by_category
```

```python
plt.pie(sales_by_category['Sales'],labels = sales_by_category['Category'],autopct = '%1.1f%%')
plt.legend(loc='upper left')
plt.title("Sales Analysis by Category")
plt.show()
```

The sales analysis done based on sub-categories

```python
sales_by_subcategory = data.groupby('Sub-Category')['Sales'].sum().reset_index()
```

```python
sales_by_subcategory
```

```python
plt.bar(sales_by_subcategory['Sub-Category'],sales_by_subcategory['Sales'],width = 0.4)
plt.xticks(rotation=70)
plt.xlabel("Sub-Category")
plt.grid(alpha = 0.2)
plt.ylabel("Sales")
plt.title("Sales Analysis by Sub Category")
plt.show()
```

# analyze the monthly profit from sales and determine which month had the highest profit

```python
profit_by_month = data.groupby('Order Month')["Profit"].sum().reset_index()
```

```python
profit_by_month
```

```python
X_labels = ['January', 'February', 'March', 'April', 'May', 'June', 
'July', 'August', 'September', 'October', 'November', 'December']
plt.bar(profit_by_month['Order Month'],profit_by_month['Profit'],width = 0.4,color = "Black")
plt.xticks(ticks = profit_by_month['Order Month'],labels = X_labels,rotation=70)
plt.xlabel("Order Month")
plt.grid(alpha = 0.2)
plt.ylabel("Profit")
plt.title("Profit Analysis by Month")
plt.show()
```

Analyze the profit by category and sub-category

```python
profit_by_category = data.groupby('Category')['Profit'].sum().reset_index()
```

```python
profit_by_category
```

```python
plt.pie(profit_by_category['Profit'],labels = profit_by_category['Category'],autopct = '%1.1f%%',colors = ['Brown','Yellow','Cyan'])
plt.legend(loc='upper left')
plt.title("Profit Analysis by Category")
plt.show()
```

```python
profit_by_subcategory = data.groupby('Sub-Category')['Profit'].sum().reset_index()
```

```python
profit_by_subcategory
```

```python
plt.bar(profit_by_subcategory['Sub-Category'],profit_by_subcategory['Profit'],width = 0.4,color = "Brown")
plt.xticks(rotation=70)
plt.xlabel("Sub-Category")
plt.grid(alpha = 0.2)
plt.ylabel("Profit")
plt.title("Profit Analysis by Sub-Category")
plt.show()
```

Analyze the sales and profit by customer segment

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
```

```python
sales_profit_by_segment
```

```python
bar_width = 0.35
x = np.arange(len(sales_profit_by_segment['Segment']))
plt.bar(x - (bar_width/2),sales_profit_by_segment['Sales'],bar_width,label = 'Sales',color = 'skyblue')
plt.bar(x + (bar_width/2),sales_profit_by_segment['Profit'],bar_width,label = 'Profit',color = 'lightgreen')
plt.xlabel('Segment')
plt.ylabel('Amount')
plt.title('Sales and Profit Analysis by Customer Segment')
plt.xticks(x, sales_profit_by_segment['Segment'])
plt.legend()
plt.grid(alpha = 0.3)
plt.show()
```

Analyze the sales to profit ratio

```python
sales_profit_by_segment['Sales_to_Profit_Ratio'] = sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment','Sales_to_Profit_Ratio']])
```
